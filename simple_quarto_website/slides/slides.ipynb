## Basic Information 
<p>Theory:</p>
<p>The binomial distribution describes the number of successes in a fixed number of independent trials, each with the same probability of success. It is a discrete distribution and is commonly used to model scenarios like flipping a coin or answering multiple-choice questions.</p>

## Equations and Graphs
Probability Mass Function:
$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

An example of binomial distribution plot:

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
n, p = 10, 0.5  # Number of trials and probability of success
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

plt.bar(x, y)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```
